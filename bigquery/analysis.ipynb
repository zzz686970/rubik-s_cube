```python
import folium
import pandas as pd
from folium import plugins
```

```python
df = pd.read_csv('users.csv')
df.head()
```

```python
df['fullVisitorId'].nunique()
## 10058
```

```python
map2 = folium.Map(location=[df['lat'].median(),df['lon'].median()],  zoom_start=6)
# Plotting 
points = df[['lat','lon']].drop_duplicates().to_records(index=False).tolist()
map2.add_child(plugins.MarkerCluster(locations=points[:]))
map2  
```
